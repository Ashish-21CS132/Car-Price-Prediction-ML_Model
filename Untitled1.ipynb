```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dataset=pd.read_csv("car_pred_price.csv")
dataset.head()
```

# cleaning

```python
dataset["year"].unique()
```

```python
dataset=dataset[dataset["year"].str.isnumeric()]
```

```python
dataset["year"]=dataset["year"].astype("int64")
dataset["year"].info()
```

```python
dataset.shape
```

```python
dataset=dataset[dataset["Price"]!="Ask For Price"]
dataset["Price"].unique()
```

```python
dataset["Price"]=dataset["Price"].str.replace(',','').astype("int64")
dataset["Price"]
```

```python
dataset["kms_driven"]=dataset["kms_driven"].str.split(' ').str.get(0).str.replace(',','')
dataset["kms_driven"]
```

```python
dataset=dataset[dataset["kms_driven"].str.isnumeric()]
```

```python
dataset["kms_driven"]=dataset["kms_driven"].astype("int64")
dataset["kms_driven"].info()
```

```python
dataset["fuel_type"].unique()
```

```python
dataset=dataset[~dataset["fuel_type"].isna()]
dataset["fuel_type"].unique()
```

```python
dataset.head()
```

```python
dataset["name"]=dataset["name"].str.split(" ").str.slice(start=0,stop=3).str.join(" ")
dataset["name"].head()
```

```python
dataset
```

```python
dataset=dataset.reset_index(drop=True)
```

```python
dataset.shape
```

```python
dataset=dataset[dataset["Price"]<6e6]
dataset.shape
```

```python
dataset["company"].unique()
dataset=dataset.reset_index(drop=True)
```

# graph company vs price

```python
plt.subplots(figsize=(15,7))
ax=sns.boxplot(x='company',y='Price',data=dataset)
ax.set_xticklabels(ax.get_xticklabels(),rotation=40,ha='right')
plt.show()
```

```python
x=dataset.drop(columns="Price")
y=dataset["Price"]
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import make_column_transformer
from sklearn.metrics import r2_score
from sklearn.pipeline import make_pipeline
```

# creating a hotOnecodeing

```python
ohe=OneHotEncoder()
ohe.fit(x[['name','company','fuel_type']])
```

# creating a columns transformers

```python
columns_transformer=make_column_transformer((OneHotEncoder(categories=ohe.categories_),['name','company','fuel_type']),remainder='passthrough')
```

```python
lr=LinearRegression()
```

# Making a pipeline

```python
pipe=make_pipeline(columns_transformer,lr)
```

# fitting the model into pipe

```python
pipe.fit(x_train,y_train)
```

```python
y_pred=pipe.predict(x_test)
```

# check r2 score

```python
r2_score(y_test,y_pred)
```
